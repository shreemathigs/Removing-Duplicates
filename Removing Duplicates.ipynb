```python
nan_rows=df[df.isna().any(axis=1)] #Returns all the missing values in rows
nan_rows
```

```python
nan_specific_rows=df[df.deck.isna()] #Shows the missing value for deck
nan_specific_rows
```

```python
duplicate_rows=df[df.duplicated()] #Shows the duplicate rows
duplicate_rows
```

```python
df = df.dropna(subset=['age', 'deck','embarked','embark_town']) #Drops the missing value rows in deck,age,embarked,embark_town
df
```
